```python
import numpy as np
from PIL import Image
from pickle import load
import matplotlib.pyplot as plt
from keras.models import load_model
from utils.model import CNNModel, generate_caption_beam_search
from utils.preprocessing import extract_features_for_test
from utils.load_data import load_doc,load_descriptions
import os
import nltk
from config import config

"""
    *Some simple checking
"""
assert type(config['max_length']) is int, 'Please provide an integer value for `max_length` parameter in config.py file'
assert type(config['beam_search_k']) is int, 'Please provide an integer value for `beam_search_k` parameter in config.py file'
```

```python

# Load the tokenizer
tokenizer_path = config['tokenizer_path']
tokenizer = load(open(tokenizer_path, 'rb'))

# Max sequence length (from training)
max_length = config['max_length']

# Load the model
caption_model = load_model(config['model_load_path'])

image_model = CNNModel(config['model_type'])

```

```python
###    Test image & caption should be this form                 ###

###    000001.jpeg baby drinking juice pack                     ###
###    000002.jpeg four boy playing ball on green grass         ###
###    000003.jpeg boy sitting on bench while holding a book    ###

doc = load_doc('test_data/smart_eye_inference_data/Smart_Eye_Inference_Data_Captions.txt')
descriptions = load_descriptions(doc)
```

```python
image_name ='000017'
image_file = image_name + '.jpeg'
image = extract_features_for_test(config['test_data_path']+image_file, image_model, config['model_type'])
# Generate caption using Decoder RNN Model + BEAM search
generated_caption = generate_caption_beam_search(caption_model, tokenizer, image, max_length, beam_index=config['beam_search_k'])
# Remove startseq and endseq
caption = generated_caption.split()[1].capitalize()
hy = []
refer = list(descriptions[image_name][0].split())
#refer = [['man', 'in', 'a', 'red', 'shirt', 'is', 'standing', 'on', 'the', 'edge', 'of', 'a', 'large', 'rock']]
for x in generated_caption.split()[2:len(generated_caption.split())-1]:
    caption = caption + ' ' + x
    hy.append(x)
caption += '.'
# Show image and its caption
x=plt.imread(config['test_data_path']+image_file)
plt.imshow(x)
plt.show()
print('Prediction: ',caption)
print('Answer: ',' '.join(hy))
#print('BLEU SCORE:',nltk.translate.bleu_score.sentence_bleu(refer,hy)*100)
```
